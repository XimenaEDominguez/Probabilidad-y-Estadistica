## Ejemplo 3: Histogramas de 2 grupos

1- En este ejemplo generamos 100 numeros aleatorios reales que representa la produccion diaria de 100 telares(medidas en metros)

2- Generar otra lista con la produccion de 100 telares para una segunda fabrica

3-Con estas listas de valores se pretende determinar las medidas de tendencia central, es decir, la media, la moda, la mediana, la desviacion estandar.

4- Graficar un histograma para cada fabrica


```python
from statistics import *
```

```python
# Libreria para generar numeros aleatorios
import random
```

```python
fabrica1= []
for i in range(100):
  numero_aleatorio= round (random.uniform(50,100) ,2)
  fabrica1.append(numero_aleatorio)
print(fabrica1)
```

```python
fabrica2= []
for i in range(100):
  numero_aleatorio2 = round (random.uniform(50,100) ,2)
  fabrica2.append(numero_aleatorio2)
print(fabrica2)
```

```python
#Libreria para graficar
import matplotlib.pyplot as plt
```

```python
plt.hist(fabrica1)
plt.title("Produccion diaria de la fabrica 1")
```

```python
plt.hist(fabrica2)
plt.title("Produccion diaria de la fabrica 2")
```

```python
media=mean(fabrica1)
mediana = median(fabrica1)
moda= mode(fabrica1)

print ("Valores de la fabrica 1")
print("La media es:", media)
print("La mediana es:", mediana)
print("La moda es:", moda)
```

```python
media=mean(fabrica2)
mediana = median(fabrica2)
moda= mode(fabrica2)

print ("Valores de la fabrica 2")
print("La media es:", media)
print("La mediana es:", mediana)
print("La moda es:", moda)
```

```python
#Generar 2 subplot de 1 renglon por 2 columnas para desplegar los 2 histogramas, una al lado de la otra
fig,(ax1,ax2) = plt.subplots(nrows =1,ncols =2)
```

```python
#Generar 2 subplot de 1 renglon por 2 columnas para desplegar los 2 histogramas, una al lado de la otra
fig,(ax1,ax2) = plt.subplots(nrows =1,ncols =2, figsize=(10,5))
ax1.hist(fabrica1)
ax2.hist(fabrica2)
plt.show()
```

```python
#Valor minimo, maximo, rango

media1 = mean(fabrica1)
mediana1 = median(fabrica1)
moda1 = mode(fabrica1)
```

```python
plt.hist (fabrica1, bins=10, alpha=0.2,color='b', edgecolor='black')
plt.axvline(media1, color='b', linestyle='dashed', linewidth=2, label='Media' )
plt.axvline(mediana1, color='r', linestyle='dashed', linewidth=2, label='Mediana' )
plt.axvline(moda1, color='y', linestyle='dashed', linewidth=2, label='Moda' )
plt.xlabel('Valor')
plt.ylabel('frecuencia')
plt.title("Produccin diaria de la fabrica1")
plt.legend()
plt.show()
```

```python
#Valor minimo, maximo, rango

media2 = mean(fabrica2)
mediana2 = median(fabrica2)
moda2 = mode(fabrica2)
```

```python
plt.hist (fabrica2, bins=20, alpha=0.2,color='b', edgecolor='black')
plt.axvline(media2, color='b', linestyle='dashed', linewidth=2, label='Media' )
plt.axvline(mediana2, color='r', linestyle='dashed', linewidth=2, label='Mediana' )
plt.axvline(moda2, color='y', linestyle='dashed', linewidth=2, label='Moda' )
plt.xlabel('Valor')
plt.ylabel('frecuencia')
plt.title("Produccin diaria de la fabrica2")
plt.legend()
plt.show()
```

## Mximo, Mnimo y Rango

```python
maximo = max(fabrica1)
minimo = min(fabrica1)
rango = maximo - minimo
media = mean(fabrica1)
mediana = median(fabrica1)
moda = mode(fabrica1)
print("Estadistica descriptiva de la fabrica1", "\nmedia =", media, "\nmediana =", mediana, "\nmoda =", moda, "\nMximo =", maximo, "\nMnimo =", minimo, "\nRango =", rango)
```

```python
maximo2 = max(fabrica2)
minimo2 = min(fabrica2)
rango2 = maximo2 - minimo2
media2 = mean(fabrica2)
mediana2 = median(fabrica2)
moda2 = mode(fabrica2)
print("Estadistica descriptiva de la fabrica2", "\nmedia =", media2, "\nmediana =", mediana2, "\nmoda =", moda2, "\nMximo =", maximo2, "\nMnimo =", minimo2, "\nRango =", rango2)
```

## Dispersin de los datos

```python
desviacion_estandar_fabrica1 = stdev(fabrica1)
print("La desviacin estandar de la fabrica 1 es = ", desviacion_estandar_fabrica1 )
```

```python
desviacion_estandar_fabrica2 = stdev(fabrica2)
print("La desviacin estandar de la fabrica 2 es = ", desviacion_estandar_fabrica2 )
```

```python
plt.figure(figsize = (10,8))
plt.hist (fabrica1, bins=10, alpha=0.2,color='b', edgecolor='black')
plt.axvline(media1, color='b', linestyle='dashed', linewidth=2, label=f'media={media1: .2f}')
plt.axvline(mediana1, color='r', linestyle='dashed', linewidth=2, label=f'Mediana={mediana1:.2f}')
plt.axvline(moda1, color='y', linestyle='dashed', linewidth=2, label=f'Moda={moda1:.2f}' )
plt.axvline(media-desviacion_estandar_fabrica1, color='black', linestyle='dashed', linewidth=2, label=f'Desv. Estndar = {desviacion_estandar_fabrica1:.2f}' )
plt.axvline(media+desviacion_estandar_fabrica1, color='black', linestyle='dashed', linewidth=2, label=f'Desv. Estndar= {desviacion_estandar_fabrica1:.2f}' )
plt.xlabel('Valor')
plt.ylabel('frecuencia')
plt.title("Produccin diaria de la fabrica1")
plt.legend()
plt.show()
```
