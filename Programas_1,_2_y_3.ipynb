```python
import math as mt
from itertools import combinations
from itertools import combinations_with_replacement
import numpy as np
```

```python
nombres = ["Valeria","Alondra", "Pedro", "Silvana", "Rebeca", "Carlos", "Lucio", "Sofia", "Kevin", "Johan"]
n = len (nombres)
r = 3
Cn = mt. factorial(n) / (mt. factorial(r) * mt. factorial(n-r))
print ("Existen", Cn ,  " nombres "" posibles combinaciones sin repetición del total de ",n," en grupos de ", r," en ", r)
print ("Aquí las combinaciones")
combinaciones_nombres = list(combinations (nombres, r))
for combinacion in combinaciones_nombres:
 print(combinacion)
```
